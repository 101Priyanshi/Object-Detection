### Importing Libraries

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

### Opening Image

```python
img = cv2.imread('Images/car_plate.jpg')
```

### Converting Image

```python
def display(img):
    fig = plt.figure(figsize=(10,8))
    ax = fig.add_subplot(111)
    new_img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
    ax.imshow(new_img)
```

### Displaying it

```python
display(img)
```

### Creating Classifier

```python
plate_classifier = cv2.CascadeClassifier('Haarcascades/haarcascade_plate_number.xml')
```

### Final Results

```python
def detect_plate(img):
    plate_img = img.copy()
    plate_rects = plate_classifier.detectMultiScale(plate_img,1.1,1)
    for(x,y,z,h) in plate_rects:
        cv2.rectangle(plate_img,
                     (x,y),#upper left corner
                     (x+z,y+h),#down right corner
                     (0,255,255),#red color
                     4)#thickness
    return plate_img
```

```python
result = detect_plate(img)
display(result)
```

### Import Libraries

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline #means we want to open the photo in this notebook and not on another window
```

### Open Image

```python
img = cv2.imread('Images/eye_face.jpg')
```

### Show Image

```python
plt.imshow(img)
```

### Fix Image

```python
fix_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

### Show Image

```python
plt.imshow(fix_img)
```

### Create Classifier

```python
eye_classifier = cv2.CascadeClassifier('Haarcascades/haarcascade_eye.xml')
```

### Show Results

```python
def detect_eye(fix_img):
    eye_rects = eye_classifier.detectMultiScale(fix_img)
    for(x,y,z,h) in eye_rects:
        cv2.rectangle(fix_img,
                     (x,y),#upper left corner
                     (x+z,y+h),#down right corner
                     (255,0,0),#red color
                     10)#thickness
    return fix_img
```

```python
result = detect_eye(fix_img)
plt.imshow(result)
```

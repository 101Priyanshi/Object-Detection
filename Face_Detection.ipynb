### Import Libraries

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline#means we want to open the photo in this notebook and not on snother window
```

### Create Classifier

```python
face_classifier = cv2.CascadeClassifier('Haarcascades/haarcascade_frontalface_default.xml')
```

### Open Image

```python
image = cv2.imread('Images/eye_face.jpg')
fix_img = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)#when opencv reads the image, it's BGR but when the image is captured it's RGB so we convert the bgr channel to rgb channel
plt.imshow(fix_img)
```

### Detect Faces

```python
faces = face_classifier.detectMultiScale(image, 1.3, 5)# we can change the parameters for better tuning
```

### No Faces Detected

```python
if faces is ():
    print("No faces found")
```

### It's a kind of Magic

```python
def detect_face(fix_img):
    face_rects = face_classifier.detectMultiScale(fix_img)
    for(x,y,z,h) in face_rects:
        cv2.rectangle(fix_img,
                     (x,y),#upper left corner
                     (x+z,y+h),#down right corner
                     (255,0,0),#red color
                     10)#thickness
    return fix_img
```

```python
result = detect_face(fix_img)
plt.imshow(result)
```
